Stock Portfolio Visualization
Scenario: You have daily closing prices and trading volumes for 4 stocks over 10 days.
Tasks:

Line plot: Plot closing prices for all stocks with different line styles.

Bar plot: Plot trading volume per stock on day 5.

Error bars: Add Â±2% error bars to price plots.

Stacked area plot: Show cumulative portfolio value over 10 days.

Annotations: Highlight the day with the maximum portfolio value.

Subplots: Create a 2x1 subplot figure: top = price trends, bottom = volume bar chart.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
days = np.arange(1, 11)
stocks = ["AAPL", "GOOG", "TSLA", "AMZN"]

prices = np.array([
    [100, 200, 50, 300],
    [102, 198, 52, 305],
    [101, 202, 51, 310],
    [105, 205, 53, 315],
    [107, 207, 54, 320],
    [110, 210, 55, 325],
    [108, 208, 56, 330],
    [111, 212, 57, 335],
    [113, 215, 58, 340],
    [115, 218, 60, 345]
])

volumes = np.random.randint(1000, 5000, size=(10,4))

class StockDashboard:
  def __init__(self, days, prices, volumes, stocks):
    self.days = days
    self.prices = prices
    self.volumes = volumes
    self.stocks = stocks

  def error_plot_prices(self):
    line_styles = ['-', '--', '-.', ':']
    for i, stock in enumerate(stocks):
        plt.errorbar(days, prices[:, i], yerr=0.02*prices[:, i],
                     label=stock, linestyle=line_styles[i])
    plt.title("Stock Prices Over 10 Days")
    plt.xlabel=("Days")
    plt.ylabel=("Price")
    plt.legend()
    plt.show()


  def plot_volume_bar_day5(self):
    day_index = 4
    plt.bar(stocks, volumes[day_index, :], color=['blue', 'orange', 'green', 'red'])
    plt.title("Trading Volume Per Stock on Day 5")
    plt.show()

  def plot_stacked_portfolio(self):
    cumulative_prices = np.cumsum(prices, axis=0)
    plt.stackplot(days, cumulative_prices[:, 0], cumulative_prices[:,1], cumulative_prices[:,2], cumulative_prices[:,3],
                   labels=stocks, colors=['blue', 'orange', 'lightgreen', 'red'])
    plt.title("Cumulative Stock Prices Over 10 Days")
    plt.legend()
    plt.show()

  def plot_subplots(self):
    fig, axs = plt.subplots(2, 1, figsize=(6, 5))
    line_styles = ['-', '--', '-.', ':']
    for i, stock in enumerate(stocks):
      axs[0].errorbar(days, prices[:, i], yerr=0.02*prices[:,1], label=stocks[i], linestyle=line_styles[i])
    axs[0].legend()
    width=0.2
    for i, stock in enumerate(stocks):
      axs[1].bar(days+i*width, volumes[:,i], width=width, label=stock)
    plt.tight_layout()
    axs[1].legend()
    plt.show()

dashboard = StockDashboard(days, prices, volumes, stocks)
dashboard.error_plot_prices()
dashboard.plot_volume_bar_day5()
dashboard.plot_stacked_portfolio()
dashboard.plot_subplots()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

hours = np.arange(0, 24)

temperature = 20 +10 * np.sin((hours-6)/24 *2 * np.pi) + np.random.randn(24)
humidity = 50 - 10 * np.sin((hours-6)/24 *2 * np.pi) + np.random.randn(24)
wind_speed = np.random.uniform(5,20,size=24)
rainfall = np.random.poisson(2,size=24)

class WeatherDashboard:
  def __init__(self, hours, temperature, humidity, wind_speed, rainfall):
    self.hours=hours
    self.temperature=temperature 
    self.humidity=humidity
    self.wind_speed=wind_speed
    self.rainfall=rainfall

  def plot_temperature(self):
    plt.errorbar(hours,temperature, yerr=0.05*temperature)
    plt.title("Temperature Over 24 Hours")
    plt.grid(True)
    plt.show()

  def plot_humidity_scatter(self):
    plt.scatter(hours, humidity, alpha=0.7, color='green')
    plt.title("Humidity Over 24 Hours")
    plt.grid(True)
    plt.show()

  def plot_wind_bar(self):
    wind_speed = np.arange(24)*10
    hours = [6, 12, 18]
    values = [wind_speed[h] for h in hours]    
    plt.bar(hours, values, color='orange', width=2)
    plt.grid(True)
    plt.show()

  def plot_rainfall_area(self):
    hours = np.arange(1, len(rainfall) +1)
    cumulative_rainfall = np.cumsum(rainfall, axis=0)
    plt.stackplot(hours, cumulative_rainfall)
    plt.legend()
    plt.show()

  def plot_subplots_dashboard(self):
    pass

  def plot_normalized_temperature(self):
    temp_norm=(temperature-np.min(temperature))/(np.max(temperature)-np.min(temperature))
    plt.plot(hours, temp_norm)
    plt.show()


  def plot_3d_surface(self):
    X,Y=np.meshgrid(hours,humidity)
    Z=np.tile(self.temperature, (24,1))
    ax=plt.axes(projection='3d')
    ax.plot_surface(X,Y,Z, cmap='viridis')
    plt.show()

dashboard = WeatherDashboard(hours, temperature, humidity, wind_speed, rainfall)
dashboard.plot_temperature()
dashboard.plot_humidity_scatter()
dashboard.plot_wind_bar()
dashboard.plot_rainfall_area()
dashboard.plot_subplots_dashboard()
dashboard.plot_normalized_temperature()
dashboard.plot_3d_surface()
```
